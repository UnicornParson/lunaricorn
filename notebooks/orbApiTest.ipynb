```python
import sys
import os
from pathlib import Path
from datetime import datetime
sys.path.append(os.path.dirname(os.getcwd()))
sys.path.append(Path(os.getcwd()).parent.absolute())
import lunaricorn
import lunaricorn.api.signaling as lsig
import lunaricorn.api.orb as orb
import uuid
ORB_HOST="192.168.0.18"
ORB_PORT=5558
```

```python
str(uuid.uuid1())
```

```python
# Создание объекта OrbClient
orb_client = orb.OrbClient(ORB_HOST, ORB_PORT)

# Проверка состояния клиента
print("Проверка состояния клиента:", orb_client.good())

# Пример использования методов OrbClient
try:
    # Пример создания и сохранения данных
    data = {"key": "value", "rkey" : str(uuid.uuid1())}
    new_orb_data = orb_client.create_orb_data(data)
    print("Сохраненный объект данных:", new_orb_data)
    
    # Пример получения данных по идентификатору
    fetched_orb_data = orb_client.fetch_orb_data(new_orb_data.u)
    if fetched_orb_data:
        print("Найденный объект данных:", fetched_orb_data)
    
    # Пример создания и сохранения мета-объекта
    new_orb_meta = orb_client.create_orb_meta(handle=123)
    print("Сохраненный объект мета-данных:", new_orb_meta)
    
    # Пример получения мета-объекта по идентификатору
    fetched_orb_meta = orb_client.fetch_orb_meta(new_orb_meta.id)
    if fetched_orb_meta:
        print("Найденный объект мета-данных:", fetched_orb_meta)
    
except Exception as e:
    print("Ошибка при работе с OrbClient:", str(e))
```
