```python
import requests
import random
import time

import math
playgroud_host = "192.168.0.12"
playgroud_port = 8080
def send_command(ip, port, data=None):
    if data is None:
        data = {}
    url = f"http://{ip}:{port}/command"
    payload = {"action": "left", "data": data}
    response = requests.post(url, json=payload)
    return response
```

```python
def brownian_motion_loop(host, port, steps=100, delay=0.05):
    prev_x, prev_y = 1.0, 0.0
    for _ in range(steps):
        angle_noise = random.uniform(-math.pi/2, math.pi/2) * 0.3
        prev_angle = math.atan2(prev_y, prev_x)
        new_angle = prev_angle + angle_noise
        length = random.uniform(0.5, 1.0)
        x = math.cos(new_angle) * length
        y = math.sin(new_angle) * length
        send_command(host, port, {"x": f"{x:.2f}", "y": f"{y:.2f}"})
        prev_x, prev_y = x, y
        time.sleep(delay)
```

```python
brownian_motion_loop(playgroud_host, playgroud_port)
```
